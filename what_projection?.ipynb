```python
import rasterio
import zipfile
from rasterio.io import MemoryFile

def check_tif_projection_in_zip(zip_path):
    # Open the ZIP file
    with zipfile.ZipFile(zip_path) as z:
        # List all files in the ZIP
        tif_files = [f for f in z.namelist() if f.endswith('.tif')]
        # Process each TIFF file
        for tif in tif_files:
            # Read the TIFF file as bytes
            with z.open(tif) as file:
                with MemoryFile(file.read()) as memfile:
                    with memfile.open() as dataset:
                        # Print the CRS info
                        print(f"Projection for {tif}: {dataset.crs}")
```
