```r
# planet 3 meter data, 768,768 = 256x256 pixels.This works correctly.
# each aoi is assigned a number 1-16 and the individual 16 data tables are rbindlist together to one.

aoi_utm4_wetland_fisnet <-
  1:nrow(aoi_utm4) %>% lapply(function(x){
    fishnet <- st_make_grid(aoi_utm4[x,], cellsize = c(768,768)) %>%
      st_as_sf() %>%
        st_intersection(aoi_utm4[x,]) %>%
          st_transform(st_crs(26904)) %>%
            data.table()


    fishnet$island <- aoi_utm4$island[x]
    fishnet$layer <- aoi_utm4$layer[x]

    return(fishnet)
    # print(fishnet)
    # print(class(fishnet))
    # print(st_crs(fishnet))
    # plot(fishnet)

  }) %>% rbindlist()

class(aoi_utm4_wetland_fisnet)
aoi_utm4_wetland_fisnet_sf <- st_as_sf(aoi_utm4_wetland_fisnet)
class(aoi_utm4_wetland_fisnet_sf)
st_write(aoi_utm4_wetland_fisnet_sf, "/content/sample_data/aoi_utm4_wetland_fisnet_sf.gpkg", append=TRUE)
```
