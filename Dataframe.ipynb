```python

from PIL import Image

# Directories
image_directory = '/content/drive/MyDrive/data/oahu'
label_directory = '/content/drive/MyDrive/data/ccap/oahu'

# Empty list to collect data
data = []

# Process each image and corresponding label
for filename in os.listdir(image_directory):
    if filename.endswith('.tif'):
        # Image file paths
        image_path = os.path.join(image_directory, filename)
        label_path = os.path.join(label_directory, filename)

        # Open the image and label with rasterio
        with rasterio.open(image_path) as src:
            # Get geographic coordinates (bounds)
            xmin, ymin, xmax, ymax = src.bounds

            # Get other metadata
            crs = src.crs
            tile_name = filename
            resolution = src.res
            dimensions = src.shape
            min_value, max_value = src.read(1).min(), src.read(1).max()

            # Extract class information from the label
            with rasterio.open(label_path) as lbl:
                # Assuming class is encoded in pixel values
                class_values = list(set(lbl.read(1).flatten()))

            # Append information to the list
            data.append([xmin, ymin, xmax, ymax, class_values, dimensions, resolution, crs, tile_name, min_value, max_value])

# Create DataFrame
df = pd.DataFrame(data, columns=['xmin', 'ymin', 'xmax', 'ymax', 'class', 'dimensions', 'resolution', 'crs', 'tile_name', 'min_value', 'max_value'])

# Convert DataFrame to CSV and save
csv_file = 'geospatial_data.csv'
df.to_csv(csv_file, index=False)

# Code to display the CSV in Google Colab (add this to your Colab notebook)
# from google.colab import files
# files.download(csv_file)
```
