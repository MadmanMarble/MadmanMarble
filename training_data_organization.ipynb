```python
# utm5 metadata csv writing

ccap_folder_path = "/content/data/content/exports/nwi_nlcd_filled"

ccap_metadata_list = []

for filename in os.listdir(ccap_folder_path):
  if filename.endswith(".tif"):
    filepath = os.path.join(ccap_folder_path, filename)

    with rasterio.open(filepath) as src:
      metadata = src.meta

      metadata_info = {
          'filename': filename,
          'width': src.width,
          'height': src.height,
          'crs': src.crs
      }
      ccap_metadata_list.append(metadata_info)

ccap_metadata_df = pd.DataFrame(ccap_metadata_list)

print(ccap_metadata_df)

```

```python
# Function to extract the numerical part of a filename
def extract_number(filename):
    match = re.search(r'\d+', filename)
    return int(match.group()) if match else 0

def list_and_sort_files(folder_path):
    files = os.listdir(folder_path)
    return sorted(files, key=extract_number)


folder_path_1 = '/content/image/content/exports/utm4_2018'
folder_path_2 = '/content/image/content/exports/utm4_2020'


files_folder_1 = list_and_sort_files(folder_path_1)
files_folder_2 = list_and_sort_files(folder_path_2)

# Find files in folder 1 not in folder 2
missing_in_folder_2 = set(files_folder_1) - set(files_folder_2)

# Find files in folder 2 not in folder 1
missing_in_folder_1 = set(files_folder_2) - set(files_folder_1)

# Print the results
if missing_in_folder_2:
    print("Files in folder 1 missing in folder 2:", missing_in_folder_2)
else:
    print("No files in folder 1 are missing in folder 2.")

if missing_in_folder_1:
    print("Files in folder 2 missing in folder 1:", missing_in_folder_1)
else:
    print("No files in folder 2 are missing in folder 1.")

# for file_1, file_2 in zip(files_folder_1, files_folder_2):
#     if file_1 != file_2:
#         print(f"File mismatch: {file_1} in folder 1 and {file_2} in folder 2")
```

```python
export_path = "/content/drive/MyDrive/DeepLearning_Project/trainingdata/planet128x128data/utm5/images_2018/"
tif_source = "/content/image/content/exports/utm5_2018/"

# Ensure the destination folder exists
if not os.path.exists(export_path):
    os.makedirs(export_path)

# List all .tif files in the source directory
tif_files = [f for f in os.listdir(tif_source) if f.endswith('.tif')]

# Copy each .tif file to the Google Drive destination folder
for tif_file in tif_files:
    source_path = os.path.join(tif_source, tif_file)
    destination_path = os.path.join(export_path, tif_file)
    shutil.copy(source_path, destination_path)
    print(f"Copied {tif_file} to Google Drive folder.")
```
